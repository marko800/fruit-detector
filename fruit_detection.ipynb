```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

# Play around

```python
model = YOLO("yolov8x.pt")
```

```python
results = model("data/Sample_Images/20190106_073101.jpg")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(6, 6))

axes[0].imshow(results[0].plot())
axes[1].imshow(results[0].orig_img)
plt.show()
```

```python
import os
```

```python
# Loop through all sample images:

folder_path = "data/Sample_Images"

for filename in os.listdir(folder_path):
    # Construct full file path
    file_path = os.path.join(folder_path, filename)
    
    print(f"Processing {file_path}...")

    results = model(file_path)

    detected_image = results[0].plot()

    plt.imshow(detected_image)
    plt.axis('off') 
    plt.show()
```

```python
# try with built in list acceptance:

folder_path = "data/Sample_Images"

file_path = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) ]
    
results=model(file_path)
```
